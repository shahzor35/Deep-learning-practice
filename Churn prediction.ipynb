```python
df.drop(columns = ["RowNumber", "CustomerId", "Surname"], inplace= True)
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns= ["Geography", "Gender"], drop_first = True, dtype= "int64")
```

```python
df
```

```python
X = df.drop(columns= ["Exited"])
Y = df["Exited"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_transf = scaler.fit_transform(X_train)
X_test_transf = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(11, activation = 'sigmoid', input_dim = 11))
model.add(Dense(11, activation = 'sigmoid', input_dim = 11))
model.add(Dense(1, activation = 'sigmoid', input_dim = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train_transf, Y_train, epochs = 100, batch_size= 32, validation_split = 0.2, verbose = 1)
```

```python
y_pread = model.predict(X_test_transf)
y_pread = np.where(y_pread > 0.5, 1, 0)
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_pread)
```
